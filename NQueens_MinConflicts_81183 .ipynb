### Implementation of the NQUeens Algorithm 

```python
import random

def solution_nqueens(n):
#Prints a solution for n queens 
	show(min_conflicts(list(range(n)), n), n)
 

def queen_conflicts(solution, n, col, row):
#Input: 
    #solution - temporary solution 
    #n - number of queens 
    #cow, row - coordinates of the queen
#Output: Number of conflicts for queen at point[col][row]
	total = 0
	for i in range(n):
		if i == col:
			continue
		if solution[i] == row or abs(i - col) == abs(solution[i] - row): #same row or same diagonal
			total += 1
	return total

def all_conflicts(solution, n):
#Output: Aray with size n with number of conflicts for every queen 
	return [queen_conflicts(solution, n, col, solution[col]) for col in range(n)]

def show(solution, n):
#Prints a temporary solution
	for i in range(n):
		row = ['_'] * n #all empty rows 
		for col in range(n):
			if solution[col] == n - 1 - i:
				row[col] = '*'
		print(''.join(row))

#Gives random positions to the array of queens 
def random_positions(arr, filter_, n):
	return random.choice([i for i in range(n) if filter_(arr[i])])
    
def min_conflicts(solution, n, iters=1000):
#Finds a solution with no conflicts with 1000 iterations 
	for k in range(iters):
		conflicts = all_conflicts(solution, n)
		if sum(conflicts) == 0:
			return solution
		col = random_positions(conflicts, lambda x: x > 0, n) #queens in a conflict 
		new_conflicts = [queen_conflicts(solution, n, col, row) for row in range(n)] #new conflitcs arasing from moving the queen 
		solution[col] = random_positions(new_conflicts, lambda x: x == min(new_conflicts), n)# moves the queen to the square with the minimum number of conflicts
	raise Exception("Not enough iterations")


```
